```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv("/content/Engine_data.csv")

# Rename for consistency
df.columns = [
    "engine_rpm", "lub_oil_pressure", "fuel_pressure", "coolant_pressure",
    "lub_oil_temp", "coolant_temp", "engine_condition", "oil_stress_index",
    "thermal_degradation_proxy", "pressure_temp_interaction",
    "fuel_oil_index", "engine_load_index"
]

# Drop unwanted columns
drop_cols = [
    "engine_load_index",
    "pressure_temp_interaction",
    "oil_stress_index",
    "fuel_oil_index",
    "coolant_temp"
]
df.drop(columns=drop_cols, inplace=True, errors="ignore")

# Create binary target and reason columns
df["engine_oil"] = np.where(
    (df["lub_oil_temp"] > 90) |
    (df["lub_oil_pressure"] < 2.0) |
    (df["thermal_degradation_proxy"] > 1.05) |
    (df["fuel_pressure"] > 18), 1, 0
)
df["engine_oil_health"] = np.select(
    [
        (df["lub_oil_temp"] > 90),
        (df["lub_oil_pressure"] < 2.0),
        (df["thermal_degradation_proxy"] > 1.05),
        (df["fuel_pressure"] > 18)
    ],
    [
        1 ,#"Overheated Oil",
        2 ,#"Low Oil Pressure",
        3 ,#"High Degradation Index",
        4 ,#"Fuel Pressure Stress"
    ],
    default=0 #"Normal"
)

# Save file
df.to_csv("Engine_Oil.csv", index=False)
from google.colab import files
files.download("Engine_Oil.csv")
```
